```python
import os
os.environ['PYSPARK_PYTHON'] = '/opt/conda/envs/python2/bin/python2'

from pyspark import SparkConf, SparkContext
sconf = SparkConf()
sconf.setMaster("local[*]")
sc = SparkContext(conf=sconf)
```

```python
from pyspark.sql import SQLContext
sqlCtx = SQLContext(sc) 
```

```python
data = sqlCtx.read.json('../../notebook_data/20newsgroups.labelled.json.gz')
```

```python
data.printSchema()
```

```python
data.registerTempTable('newsgroups')
docs = sqlCtx.sql("SELECT label, content FROM newsgroups")
labelsDistinct = sqlCtx.sql("SELECT DISTINCT(label) FROM newsgroups")
labelList = labelsDistinct.map(lambda r: r.label.encode('utf-8')).collect()
sc.broadcast(labelList)
```

```python
import re

labelledDocs = docs.map(lambda r: (labelList.index(r.label), r.label.encode('utf-8'), re.sub('\s+', ' ', r.content.encode('utf-8')).strip().decode('utf-8').lower()))
print labelledDocs.take(10)
print labelledDocs.count()
```

```python
from nltk.tokenize import RegexpTokenizer

tokenizer = RegexpTokenizer(r'\w+')

tokenizedDocs = labelledDocs.map(lambda (i,l,c) : (i, l, tokenizer.tokenize(c)))
tokens = tokenizedDocs.map(lambda (i,l,c) : c)
print tokens.take(2)
```

```python
from pyspark.mllib.feature import Word2Vec
import timeit

start_time = timeit.default_timer()

word2vec = Word2Vec()
model = word2vec.fit(tokens)

elapsed = timeit.default_timer() - start_time
print elapsed
```

```python
syms = model.findSynonyms("religion", 10)
print syms
```

```python
import timeit

start_time = timeit.default_timer()

test = model.getVectors()

print len(test)
print test["religion"]
elapsed = timeit.default_timer() - start_time
print elapsed
```

```python
import timeit

start_time = timeit.default_timer()

import numpy as np

vocabsize = len(test)
any_word = "religion"  
   
tmp_list = model.findSynonyms(any_word, vocabsize-1) 

list_words = []
for l in tmp_list:
    list_words.append(l[0])

list_words.append(any_word)

nwords = len(list_words)
nfeatures = model.transform(any_word).array.shape[0]

print "\n================================================="
print "Number of words in the model:", nwords
print "================================================="
print "Number of features per word: ", nfeatures
print "=================================================\n"


feature_matrix = [] 
for word in list_words:
    feature_matrix.append(model.transform(word).array)
    
elapsed = timeit.default_timer() - start_time
print elapsed
```

```python
import numpy as np
def mapWords(wordList, w2Vectors, numFeatures):
    #print len(wordList)
    numWords = len(wordList)
    #wvecs = [w2Vectors[w] for w in wordList]
    wvecs=[]
    for w in wordList:
        wvecs.append(w2Vectors.get(w, [0] * numFeatures))
    #print wvecs
    swvecs = np.sum(wvecs, axis=0)
    #print swvecs
    return [float(d)/float(numWords) for d in swvecs]


aap = mapWords(["christian", "religion"], test, nfeatures)
print "\n"
print aap
```

```python
modelVectors = {}
for w in list_words :
    modelVectors[w] = list(test[w])

print type(modelVectors["christian"])    
print modelVectors["christian"]

aap = mapWords(["christian", "religion"], modelVectors, nfeatures)
print "\n"
print aap
```

```python
docFeatures = tokenizedDocs.map(lambda (i,l,c) : (i,l,mapWords(c, modelVectors, nfeatures)))
print docFeatures.take(2)
```

```python
from pyspark.mllib.regression import LabeledPoint
from pyspark.mllib.feature import Normalizer
normalizer = Normalizer()

inputData = docFeatures.map(lambda (i,l,v): LabeledPoint(i,normalizer.transform(v)))
```

```python
print inputData.first()
training = inputData
testing = inputData
#training, testing, validation = inputData.randomSplit([0.7,0.2,0.1])
print training.first()
print testing.first()
```

```python
import timeit
start_time = timeit.default_timer()

from pyspark.mllib.classification import NaiveBayes, NaiveBayesModel
model = NaiveBayes.train(training, 1.0)

elapsed = timeit.default_timer() - start_time
print elapsed
```

```python
predictionAndLabel = testing.map(lambda p : (model.predict(p.features), p.label))
print predictionAndLabel.take(10)
accuracy = 1.0 * predictionAndLabel.filter(lambda (x, v): x == v).count() / testing.count()
print accuracy
```
